###Phase 1: Temporal-Weather Data
We merge and aggregate disease cases and weather data to a final dataset, that the observations are clusters of weeks, features are the weather condition of the week such as temperature, humidity and rainfall. Then we will bin the number of disease cases to create a categorical variable (low case, mid case and high case for example), and the target label will be this Dengue Fever outbreak risk label. Then we will split the data as training set, test set and validation set for machine learning process. Then we plan to conduct both Naive Bayes and Neural Network to compare the prediction performance.

Disease data from: https://data.gov.sg/dataset/weekly-infectious-bulletin-cases

Weather data from: https://www.ncdc.noaa.gov/cdo-web/datatools/findstation; 
                   http://www.weather.gov.sg/climate-historical-daily/


```python
dengue = pd.read_csv('/Users/ericlaibuffalo/CUSP/Data/Singapore/weekly-infectious-bulletin-cases/weekly-infectious-bulletin-cases.csv')
np.unique(dengue.disease.values)
```

```python
# Subset to get Dengue info
dengue = dengue[(dengue.disease=='Dengue Fever')|(dengue.disease=='Dengue Haemorrhagic Fever')]
dengue.reset_index(inplace=True)
```

```python
#Weather data:
dir_ = 'http://www.weather.gov.sg/climate-historical-daily/'
files_=[]
data = {}
#dir_ = 'WeatherData'
for directory,subdirectories,files in os.walk(dir_):
    for file_ in files:
        files_.append(os.path.join(dir_, file_))
        data[file_] = pd.read_csv((dir_+'/'+file_))
```
